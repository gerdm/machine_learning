```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pydataset import data
```

```python
iris = data("iris")
iris = iris.query("Species == 'setosa' | Species == 'virginica'")
```

```python
iris["Sepal"] = iris["Sepal.Length"] * iris["Sepal.Width"]
iris["Petal"] = iris["Petal.Length"] * iris["Petal.Width"]
```

```python
iris.dtypes
```

```python
%matplotlib notebook
tickx = np.linspace(iris.Sepal.min(), iris.Sepal.max(), 100)
sns.swarmplot("Sepal", "Petal", "Species", data=iris)
plt.xticks(range(5))
plt.show()
```
