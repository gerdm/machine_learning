# Convex Functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 5)
```

**1. Exercise:** Which of the following functions are convex?

a) 
$$
    f(x) = \frac{x^2 + 1}{x + 2}; \mathbf{dom}\; f=(-\infty ,-2)
$$

b) 
$$
    f(x) = \frac{1}{1-x^2}; \mathbf{dom}\; f=(-1,1)
$$

c) 
$$
    f(x) = \cosh (x) = \frac{e^x + e^{-x}}{2}
$$

```python
# a) is a concave function since it has negative curvature
xrange = np.linspace(-2.5, -2.009, 100)
plt.plot(xrange, (xrange ** 2 + 1) / (xrange + 2))
```

```python
# b) is a convex function since it has positive curvature 
xrange = np.linspace(-0.99, 0.99, 100)
plt.plot(xrange, 1 / (1 - xrange ** 2))
```

```python
# c) 
xrange = np.linspace(-np.pi, np.pi, 100)
plt.plot(xrange, np.cosh(xrange))
```

**2. Exercise**

Is 
$$
\{ x \in \mathbf{R}^n \mid \mathbf{1}^ T(x)_- \leq (1/2)\mathbf{1}^ T(x)_+\}
$$

A convex set?

```python
x1, x2 = np.mgrid[-3:3:0.01, -3:3:0.01]
x1plus, x2plus = np.maximum(0, x1), np.maximum(0, x2)
x1minus, x2minus = np.maximum(0, -x1), np.maximum(0, -x2)
```

```python
C = x1minus + x2minus <= (x1plus + x2plus) / 2
plt.contourf(x1, x2, C)
plt.colorbar()
```

## A small review on taylor expansion
Consider $f(x) = \sin(x)$ then,

```python
from scipy.misc import derivative
```

```python
@np.vectorize
def f(x): return np.sin(x)

x0 = 0
derivative(f, x0, n=3, order=5)
```

```python
x = np.linspace(-np.pi, np.pi, 100)
plt.plot(x, f(x))
plt.plot(x, f(x0) + derivative(f, x0, n=1, order=5) * (x - x0))
plt.plot(x, f(x0) + derivative(f, x0, n=1, order=5) * (x - x0)  +\
            derivative(f, x0, n=2, order=5) * (x - x0) ** 2 / 2)
plt.plot(x, f(x0) + derivative(f, x0, n=1, order=5) * (x - x0)  +\
            derivative(f, x0, n=2, order=5) * (x - x0) ** 2 / 2 +\
            derivative(f, x0, n=3, order=5) * (x - x0) ** 3 / (2 * 3))
```
