# Convex Functions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12, 5)
```

**1. Exercise:** Which of the following functions are convex?

a) 
$$
    f(x) = \frac{x^2 + 1}{x + 2}; \mathbf{dom}\; f=(-\infty ,-2)
$$

b) 
$$
    f(x) = \frac{1}{1-x^2}; \mathbf{dom}\; f=(-1,1)
$$

c) 
$$
    f(x) = \cosh (x) = \frac{e^x + e^{-x}}{2}
$$

```python
# a) is a concave function since it has negative curvature
xrange = np.linspace(-2.5, -2.009, 100)
plt.plot(xrange, (xrange ** 2 + 1) / (xrange + 2))
```

```python
# b) is a convex function since it has positive curvature 
xrange = np.linspace(-0.99, 0.99, 100)
plt.plot(xrange, 1 / (1 - xrange ** 2))
```

```python
# c) 
xrange = np.linspace(-np.pi, np.pi, 100)
plt.plot(xrange, np.cosh(xrange))
```
